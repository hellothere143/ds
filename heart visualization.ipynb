# Plotting


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
# Convert all columns to numeric
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),cmap = 'YlGnBu', annot = True)
```

# Line Chart

```python
sns.lineplot(data=df,x=df.age,y=df.cp,hue='num')
```

```python
sns.lineplot(data=df,x=df.age,y=df.trestbps,hue='num')
```

```python
sns.lineplot(data=df,x=df.age,y=df.chol,hue='num')
```

```python
sns.lineplot(df,x=df.oldpeak,y=df.exang,hue='num')
```

```python
# Shows the Distribution of Heat Diseases with respect to male and female
sns.histplot(data=df, 
             x=df.num,
             hue=df.sex)
```

```python
# Shows the Distribution of cp types with respect to male and female
sns.histplot(data=df, 
             x=df.cp,
             hue=df.sex)
```

```python
sns.histplot(data=df,x=df.cp, hue='num')
```

```python
# Shows the Distribution of age w.r.t output
sns.histplot(data=df,x=df['age'], hue='num')
```

```python
sns.histplot(data=df,x=df.trestbps, hue='num')
```

```python
temp_df = df[['age','cp', 'trestbps','chol','fbs','oldpeak','num']]
plt.figure(figsize=(15,10))
sns.pairplot(temp_df,hue="num")
plt.title("Looking for Insites in Data")
plt.legend("HeartDisease")
plt.tight_layout()
plt.plot()
```

```python

plt.figure(figsize=(15,10))
for i,col in enumerate(temp_df.columns,1):
    plt.subplot(4,3,i)
    plt.title(f"Distribution of {col} Data")
    sns.histplot(df[col],kde=True)
    plt.tight_layout()
    plt.plot()
```
